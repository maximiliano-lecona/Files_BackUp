# Archivo para rellenar los rawdata de una misma cadena

- Utilizando dos archivos de la misma cadena

```python
import pandas as pd
```

```python
archivo1 = ""
archivo2 = ""

fecha = ""
```

```python
df1 = pd.read_csv(f"Relleno/{archivo1}.csv", dtype = str, encoding='utf-8')
df2 = pd.read_csv(f"Relleno/{archivo2}.csv", dtype = str, encoding='utf-8')
print(df1.shape)
print(df2.shape)
```

```python
df1["Store ID"]
```

```python
# Limpiar ceros a la izquierda solo en columnas UPC y SKU
for col in ['UPC', 'SKU']:
    if col in df1.columns:
        df1[col] = df1[col].astype(str).str.strip().str.lstrip('0')
    if col in df2.columns:
        df2[col] = df2[col].astype(str).str.strip().str.lstrip('0')
```

```python
df_total = pd.concat([df1, df2])
df_total.shape
```

```python
df_total[['SKU','UPC','Store ID']] = df_total[['SKU','UPC','Store ID']].astype(str)  
df_total_final = df_total.drop_duplicates(subset = ['Canal','SKU', 'UPC', 'Store ID'])
df_total_final.shape
```

```python
df_total_final["Date"] = f"{fecha}"
```

```python
df_total_final["Date"].value_counts()
```

```python
new_dir = f"{archivo1}.csv"

df_total_final.to_csv(new_dir, index=False, encoding="utf-8-sig")
new_dir
```
