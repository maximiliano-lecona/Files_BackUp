# Archivo para agregar last_price a otro

- Necesita dos archivos, al que le agregaremos el last_price y el archivo que tiene los last_price

```python
import pandas as pd  
```

```python
file_last_price = ".csv"
file_price = ".csv"
```

```python
df_last_price = pd.read_csv(f"Entregas/02/{file_last_price}", dtype = str, encoding='utf-8') 
df_last_price.shape
```

```python
df_last_price.columns = df_last_price.columns.str.lower()
```

```python
df_last_price = df_last_price.drop_duplicates(
        subset=['canal','sku', 'upc'],
        keep='first'  # o 'first' según tu lógica de negocio
    )
```

```python
df_last_price.drop(columns=['last_price'], inplace=True)  # Elimina 'last_price' si existe
```

```python
df_last_price.rename(columns={'final price': 'last_price_full'}, inplace=True)  # Renombra la columna
```

```python
# Archivo al que le agregaremos el last price
dfFiltered_Data = pd.read_csv(f"Entregas/02/{file_price}", dtype = str, encoding='utf-8') 
dfFiltered_Data.shape
```

```python
dfFiltered_Data.columns = dfFiltered_Data.columns.str.lower()
```

```python
#dfFiltered_Data.drop(columns=['last_price'], inplace=True)  # Elimina 'last_price' si existe
```

```python
dfFiltered_Data = dfFiltered_Data.drop_duplicates(
        subset=['canal', 'sku', 'upc'],
        keep='first'  # o 'first' según tu lógica de negocio
    )
```

```python
dfFiltered_Data = pd.merge(dfFiltered_Data, df_last_price[['canal', 'sku', 'upc','last_price_full']], 
                    on=['canal', 'sku', 'upc'], 
                    how='left')
```

```python
dfFiltered_Data['last_price_full'] = dfFiltered_Data['last_price_full'].fillna("")
```

```python
# Ahora comparar
dfFiltered_Data['last_price'] = dfFiltered_Data.apply(
    lambda row: "" if (row['last_price_full'] == "" or row['last_price_full'] == row['final price']) else row['last_price_full'],
    axis=1
)
```

```python
dfFiltered_Data.drop(columns=['last_price_full'], inplace=True)  # Elimina 'last_price' si existe
```

```python
dfFiltered_Data.to_csv(f"{file_price}", index = False, encoding = "utf-8-sig")
```
