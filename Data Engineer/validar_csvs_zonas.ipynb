# Codigo para validar todos los archivos de 7eleven Didi

```python
import pandas as pd
import os

# Configuración
doc_name = "7-eleven"
fecha = ""
carpeta = "./csvs_zonas/"

# Store IDs esperados por zona
ZONAS = {
    "nuevo_leon": ['64000_didi_oxxo', '67174_didi_oxxo_nuevoleon', '66278_didi_oxxo_nuevoleon', '64650_didi_oxxo', '66630_didi_oxxo_nuevoleon', '66054_didi_oxxo_nuevoleon'],
    "tijuana": ['22010_didi_oxxo', '22024_didi_oxxo', '22400_didi_oxxo',],
    "rosarito": ['22707_didi_oxxo', '22706_didi_oxxo', '22703_didi_oxxo'],
    "ensenada": ['22890_didi_oxxo', '22840_didi_oxxo', '22830_didi_oxxo'],
    "hermosillo": ['83190_didi_oxxo', '83200_didi_oxxo', '83110_didi_oxxo', '83220_didi_oxxo'],
    "guadalajara": ['44610_didi_oxxo', '44100_didi_oxxo', '44700_didi_oxxo', '44840_didi_oxxo', '44648_didi_oxxo'],
    "merida": ['97160_didi_oxxo', '97000_didi_oxxo', '97300_didi_oxxo', '97070_didi_oxxo'],
    "ciudad_de_mexico": ['09000_didi_oxxo', '11000_didi_oxxo', '03000_didi_oxxo', '53150_didi_oxxo', '55107_didi_oxxo', '54030_didi_oxxo'],
    "playa_del_carmen": ['77713_didi_oxxo', '77720_didi_oxxo', '77710_didi_oxxo'],
    "cancun": ['77536_didi_oxxo', '77500_didi_oxxo', '77560_didi_oxxo']
}


def validar_todos():
    print(f"\nVALIDACIÓN DE STORE IDs - {fecha}\n")
    
    for zona, lst_store_ids in ZONAS.items():
        archivo = f"{carpeta}{doc_name}_{zona}_{fecha}.csv"
        
        if not os.path.exists(archivo):
            print(f"❌ {zona}: Archivo no encontrado")
            continue
        
        df = pd.read_csv(archivo, dtype=str, encoding='utf-8-sig')
        df_store_ids = set(df['store id'].unique())
        missing_ids_in_df = list(set(lst_store_ids) - df_store_ids)
        
        if missing_ids_in_df:
            print(f"❌ {zona}: Faltan {len(missing_ids_in_df)} store IDs -> {missing_ids_in_df}")
        else:
            print(f"✅ {zona}: Todos los store IDs presentes ({len(lst_store_ids)})")


if __name__ == "__main__":
    validar_todos()
```
