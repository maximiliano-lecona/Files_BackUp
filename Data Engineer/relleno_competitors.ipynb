# Archivo para rellenar los productos de multiples store ids. 

- Utilizando dos archivos del mismo canal

```python
import pandas as pd
```

```python
nuevo_nombre = ""
```

```python
df1 = pd.read_csv(f".csv", dtype = str, encoding='utf-8')
print(df1.shape)
df2 = pd.read_csv(f".csv", dtype = str, encoding='utf-8')
print(df2.shape)
```

```python
df1.columns
```

```python
ids = [
]

df1_bodega = df1[df1["store id"].isin(ids)]
df1_bodega['store id'].value_counts()
```

```python
df2_bodega = df2[df2["store id"].isin(ids)]
df2_bodega['store id'].value_counts()
```

```python
df_bodega_total = pd.concat([df1_bodega, df2_bodega ])
df_bodega_total.shape
```

```python
df_bodega_total['store id'].value_counts()
```

```python
df_bodega_total[['sku','upc','store id']] = df_bodega_total[['sku','upc','store id']].astype(str)

#for col in ['Canal', 'SKU', 'UPC', 'Store ID']:
#    df_bodega_total[col] = df_bodega_total[col].astype(str).str.strip().str.upper()


df_bodega_total_final = df_bodega_total.drop_duplicates(subset = ['canal','sku', 'upc', 'store id'])
df_bodega_total_final['store id'].value_counts()
```

```python
df1_sin_bodega = df1[~df1["store id"].isin(ids)]
```

```python
df_final = pd.concat([df1_sin_bodega, df_bodega_total_final])
df_final.shape
```

```python
df_final["date"] = ""
```

```python
new_dir = f"{nuevo_nombre}.csv"
df_final.to_csv(new_dir, index=False, encoding="utf-8-sig")
```
